- Use nps_chat and wordlist corpus.
- If you find a word in nps_chat but not in wordlist, it is considered an unusual word. (e.g. Heyyyy, gr8, 4ever can be - considered unusual words for this analysis)
- Find out how many unusual words people speak in chat (notice: usernames should not be counted)
- Bonus: What if I want to  count : ) , : ( all these emoticons in. (optional)


```python
from nltk.corpus import nps_chat
```

```python
from nltk.corpus import words 
```

```python
wordList = words.words()
```

```python
#L = nps_chat.posts()
L = nps_chat.fileids()
print (L)
```

```python
unusualList = []
for content in L:
    for list in nps_chat.posts(content):
        for word in list:
            if word.lower() not in wordList and type(word) != int and word.startswith('U') is False and word != "," and word != "." and word != ":" and word != "?" and word != "!" and word != ";" and word != "'" and word != "-" and word != "/" and word != "..." :
                unusualList.append(word)
            
print('The number of unusual words is ',len(unusualList))
```
