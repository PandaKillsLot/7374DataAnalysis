```python
from nltk.corpus import gutenberg
gutenberg.fileids()
```

```python
txt = gutenberg.words('austen-persuasion.txt')
# turn into lower case
txtLow = [word.lower() for word in txt]
for word in txtLow:
    if word == "," or word == "." or word == ":" or word == "?" or word == "!" or word == ";" :
        txtLow.remove(word)

# use sets to remove duplicate words
txtUni = set(txtLow)

print("1. There are " , len(txtLow), " words in austen-persuasion.txt")
print("2. There are " , len(txtUni), " unique words in austen-persuasion.txt")
```

```python
result = {}
for word in txtLow:  
    if word not in result:  
        result[word] = 0  
    else :
        result[word] += 1   
print(result)  
```
