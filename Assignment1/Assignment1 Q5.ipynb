```python
from nltk.corpus import gutenberg
import collections
```

```python
txt = gutenberg.words('shakespeare-caesar.txt') + gutenberg.words('shakespeare-hamlet.txt') + gutenberg.words('shakespeare-macbeth.txt')
# turn into lower case
txtLow = [word.lower() for word in txt]
for word in txtLow:
    if word == "," or word == "." or word == ":" or word == "?" or word == "!" or word == ";" or word == "'" or word == "-":
        txtLow.remove(word)

# use sets to remove duplicate words
txtUni = set(txtLow)
```

```python
result = {}
for word in txtLow:  
    if word not in result:  
        result[word] = 0  
    else :
        result[word] += 1   
print(result)  
```

```python
orderedResult = collections.OrderedDict(sorted(result.items(), key = lambda t: -t[1])) 
L = list(orderedResult.keys())
ska= (L[0:50])
print(ska)
```
