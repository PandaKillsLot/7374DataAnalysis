```python
#find arthors in a particular month
def findMonthlyAurthor(jsonfilepath):
    P = json.load(open( jsonfilepath))
    nameList = []
    n = 0
    for x in P["response"]["docs"]:
        n = n + 1
        if (x["byline"] is None) or (type(x["byline"]) is list) :
            name = "unknown"
        else:
            name = x["byline"]["original"]
        nameList.append(name)
        return nameList
```

```python
#find arthors in a whole year
def findYearAurthor(jsonfilepath):
    L = []
    for x in glob(jsonfilepath + "/*"):
        L = L + findMonthlyAurthor(x)
        
    fd = FreqDist(L)
    #print('Number of NAME in L:\n>>', fd.N())   
    #print('Unique NAME in l:\n>>', fd.B())
    word_list = []
    for word in fd.keys():
        word_list.append((word, fd[word]))
    sorted_list = sorted(word_list, key=lambda t:t[1], reverse=True)
    #print('Top ten aurthors in the year', sorted_list[0:10])
    if sorted_list[0][0] == "unknown":
        return sorted_list[1]
    else:
        return sorted_list[0]
```

```python
# get all the single words
def getSingleKeywords(jsonfilepath):
    P = json.load(open( jsonfilepath))
    sentList = []
    n = 0
    for x in P["response"]["docs"]:
        n = n + 1
        if len(x["keywords"]) == 0 :
            sent = "none"
        else:
            sent = x["keywords"][0]["value"]
            sentList.append(sent)
    return sentList
```

```python
#find the top ten key words
def findTopTenWords(arr):
    fd = FreqDist(arr)
    #print('Number of NAME in L:\n>>', fd.N())   
    #print('Unique NAME in l:\n>>', fd.B())
    word_list = []
    for word in fd.keys():
        word_list.append((word, fd[word]))
    sorted_list = sorted(word_list, key=lambda t:t[1], reverse=True)
    L = sorted_list[0:10]
    return L
```

```python
# find the top words for every month between year 2015 ~ 2016, and store it in a dic like {month:[top words] }
def getAllKeywords(wholefilepath):
    allTopWords = []
    yearFolder = glob(wholefilepath + "/*")
    for singleyear in yearFolder:
        fileList = glob(singleyear + "/*")
        for x in fileList:
            L = findTopTenWords(getSingleKeywords(x))
            month = x.split("/")[-1]
            dic ={}
            dic[month] = L
            allTopWords.append(dic)
    return allTopWords
            
```
