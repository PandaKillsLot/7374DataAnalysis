# Question 2

- Read all the json files in the folder called Data.
* Read only the json files which contain the key called “restaurants”
*	Each (or most of the json files) contain a key called “open” which contains the details of the operation (timings) of the restaurants. For each json file, read the timings of the restaurants. 
*	Data of the operation timings of the restaurants is present for each day of the week. I want you to extract each of this data and write it in an excel sheet. 
*	Example: 
	For a particular restaurant named “The Coffee Grounds”, the excel sheet should look like this:


```python
import csv
```

```python
# seperate the data according to their terms, if term is null, store the value as unknown
def readAndWrite (data):
    L = []
    for y in data:
        P = json.load(open(y))
        if P["term"] == "restaurants":
            #print(y)
            name = P["name"]
            city = P["location"]["city"]
            country = P["location"]["country"]
            #print(P["hours"])
            if "hours" in P:
                for x in P["hours"][0]["open"]:
                    detail = []
                    detail.append(name)
                    detail.append(city)
                    detail.append(country)
                    detail.append(x["day"])
                    detail.append(x["start"][0:2])
                    detail.append(x["start"][2:4])
                    detail.append(x["end"][0:2])
                    detail.append(x["end"][2:4])
                    L.append(detail)
            else:
                detail = []
                detail.append(name)
                detail.append(city)
                detail.append(country)
                detail.append("Unknown")
                detail.append("Unknown")
                detail.append("Unknown")
                detail.append("Unknown")
                detail.append("Unknown")
                L.append(detail)
    return L
```

```python
#insert the title
L =  readAndWrite(data)
title = ["name of the restaurant","country","city","days of the week","start time-hours","start time-minutes","end time-hours","end time-minutes"]
L.insert(0,title)
print(L[0:10])
```

```python
# write it as a csv file
out = open("RESTAURANT.csv", 'w',errors = 'ignore')
csv_writer = csv.writer(out)
for x in L:
    csv_writer.writerow(x)
out.close()
```
